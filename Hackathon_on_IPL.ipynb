**Q)  To plot the most consistent batsmen among top 10 run getters, what plotting technique we use?**

```python
top_10_batsmen = merged_df.groupby('batsman')['total_runs'].sum().nlargest(10).reset_index()

# Plotting the consistency of top 10 run-getters using a line plot
plt.figure(figsize=(10, 6))
for index, row in top_10_batsmen.iterrows():
    batsman_data = merged_df[merged_df['batsman'] == row['batsman']]
    runs_per_match = batsman_data.groupby('match_id')['batsman_runs'].sum().cumsum()
    matches_played = runs_per_match.index
    plt.plot(matches_played, runs_per_match, marker='o', label=row['batsman'])

plt.xlabel('Matches Played')
plt.ylabel('Total Runs')
plt.title('Consistency of Top 10 Run-Getters')
plt.legend()
plt.grid(True)
plt.show()
```

**Q)  In 2019 which batsman scored highest number of runs by hitting 6's and 4's?**

**Q) What is the strike rate of Kohli in 2016 ?**

```python
kohli_2016 = merged_df[(merged_df['batsman'] == 'Virat Kohli') & (merged_df['season'] == 2016)]

# Total runs and total balls faced by Kohli in 2016
total_runs_kohli = kohli_2016['batsman_runs'].sum()
total_balls_faced_kohli = len(kohli_2016)

# Calculating the strike rate
if total_balls_faced_kohli > 0:
    strike_rate_kohli = (total_runs_kohli / total_balls_faced_kohli) * 100
    print("Virat Kohli's strike rate in 2016:", round(strike_rate_kohli, 2))
else:
    print("Virat Kohli did not face any balls in 2016.")

    
```

## Insights and Observations:
- Matches Per Year: The number of matches played each year shows an increasing trend until a certain point, indicating the growth and popularity of the IPL. After reaching a peak, the number remains relatively stable.

- Venue Hosting Matches: Eden Gardens appears to have hosted the most number of IPL matches, indicating its significance as a venue for the tournament.

- Toss Decisions: Teams seem to prefer fielding first slightly more than batting first after winning the toss, as observed from the toss decisions made across matches.

- Matches Won based on Toss Decision: There seems to be a fairly balanced outcome between teams winning matches when choosing to bat first and field first after winning the toss.

- Batsmen with Most Run-Outs: The top batsmen involved in run-outs across seasons might indicate their aggressive running between the wickets or instances of misunderstanding while running.

- Hard-Hitting Ability: Players with the most number of boundaries (4s and 6s) tend to showcase a higher hard-hitting ability, crucial for scoring more runs in limited-overs cricket.

- Dismissals via Caught and Bowled: Some players are consistently getting dismissed via 'caught and bowled,' possibly suggesting either their aggressive batting style or the bowlers' ability to induce such dismissals.
